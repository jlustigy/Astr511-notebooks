## I. Getting the data

```python
from astroML import datasets
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
plt.rcParams['text.usetex']=True
import pyfits
%matplotlib inline
```

#### Load Stars from astroML

```python
# Stars from astroML
data_stars = datasets.fetch_sdss_S82standards(crossmatch_2mass=True)
sbands = ['mmed_u','mmed_g','mmed_r','mmed_i','mmed_z','J','H','K']

stars_g_r = data_stars['mmed_g'] -  data_stars['mmed_r']
stars_r_i = data_stars['mmed_r'] - data_stars['mmed_i']
stars_g_i = data_stars['mmed_g'] -  data_stars['mmed_r']
stars_J_K = data_stars['J'] -  data_stars['K']
stars_H_K = data_stars['H'] -  data_stars['K']
stars_J_H = data_stars['J'] -  data_stars['H']
stars_i_K = data_stars['mmed_i'] -  data_stars['K']
```

#### Load Quasars from astroML

```python
# Quasars from astroML
data_qsr = datasets.fetch_dr7_quasar()
q_bands = ['mag_u','mag_g','mag_r','mag_i','mag_z','mag_J','mag_H','mag_K']
qsr_redshift = data_qsr['redshift']
qsr_redshift_cut = qsr_redshift.compress((qsr_redshift < 2.5))

qsr_g_r = (data_qsr['mag_g'] - data_qsr['mag_r']).compress((qsr_redshift < 2.5))
qsr_r_i = (data_qsr['mag_r'] - data_qsr['mag_i']).compress((qsr_redshift < 2.5))
qsr_g_i = (data_qsr['mag_g'] -  data_qsr['mag_r']).compress((qsr_redshift < 2.5))
qsr_J_K = (data_qsr['mag_J'] -  data_qsr['mag_K']).compress((qsr_redshift < 2.5))
qsr_H_K = (data_qsr['mag_H'] -  data_qsr['mag_K']).compress((qsr_redshift < 2.5))
qsr_J_H = (data_qsr['mag_J'] -  data_qsr['mag_H']).compress((qsr_redshift < 2.5))
qsr_i_K = (data_qsr['mag_i'] -  data_qsr['mag_K']).compress((qsr_redshift < 2.5))
```

#### Load Galaxies from NYU Value-Added Galaxy Catalog

```python
data_gal_2mass = pyfits.getdata('/astro/users/jlustigy/Desktop/object_twomass.fits',0)
data_gal_sdss = pyfits.getdata('/astro/users/jlustigy/Desktop/object_sdss_imaging.fits',0)
gal_sdss_nMgy = data_gal_sdss['PSFFLUX'] # Units: nano-maggies!
```

Conversion from nano-maggies to magnitude:
$m = [22.5 \text{mag}] - 2.5 \text{log}_{10}f$

```python
gal_sdss_mag = 22.5 - 2.5*np.log10(gal_sdss_nMgy.T) #[u,g,r,i,z]

gal_g_r = (gal_sdss_mag[1] - gal_sdss_mag[2])
gal_r_i = (gal_sdss_mag[2] - gal_sdss_mag[3])
gal_g_i = (gal_sdss_mag[1] - gal_sdss_mag[3])
gal_J_K = (data_gal_2mass['J_M_K20FE'] - data_gal_2mass['K_M_K20FE'])
gal_H_K = (data_gal_2mass['H_M_K20FE'] - data_gal_2mass['K_M_K20FE'])
gal_J_H = (data_gal_2mass['J_M_K20FE'] - data_gal_2mass['H_M_K20FE'])
gal_i_K = (gal_sdss_mag[3] - data_gal_2mass['K_M_K20FE'])

nonzero = (gal_H_K != 0)*(gal_J_H != 0)
```

## II. Making the Firgure

```python
x0 = 300
x1 = 100
x2 = 2000
a = 0.3

fig = plt.figure(figsize=(12,12))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(stars_g_r[::x0],stars_r_i[::x0],alpha=a, color='blue', label='Stars')
ax1.scatter(qsr_g_r[::x1],qsr_r_i[::x1], alpha=a, color='green', label='QSOs')
ax1.scatter(gal_g_r[::x2],gal_r_i[::x2], alpha=a, color='red', label='Galaxies')
ax1.set_xlabel('$g - r$',fontsize=18)
ax1.set_ylabel('$r - i$',fontsize=18)
ax1.tick_params(labelsize=16)
ax1.axis([-0.5,2.2,-0.5,2.2])
ax1.legend(loc=2)

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(stars_g_i[::x0],stars_J_K[::x0],alpha=a, color='blue')
ax2.scatter(qsr_g_i[::x1],qsr_J_K[::x1], alpha=a, color='green')
ax2.scatter(gal_g_i[::x2],gal_J_K[::x2], alpha=a, color='red')
ax2.set_xlabel('$g - i$',fontsize=18)
ax2.set_ylabel('$J - K_{s}$',fontsize=18)
ax2.tick_params(labelsize=16)
ax2.axis([-1.,4.0,0.1,2.5])

ax3 = fig.add_subplot(2,2,3)
ax3.scatter(stars_J_H[::x0],stars_H_K[::x0],alpha=a, color='blue')
ax3.scatter(qsr_J_H[::x1],qsr_H_K[::x1], alpha=a, color='green')
ax3.scatter(gal_J_H[nonzero][::x2],gal_H_K[nonzero][::x2], alpha=a, color='red')
ax3.set_xlabel('$J - H$',fontsize=18)
ax3.set_ylabel('$H - K_{s}$',fontsize=18)
ax3.tick_params(labelsize=16)
ax3.axis([0.0,1.1,0.0,1.1])

ax4 = fig.add_subplot(2,2,4)
ax4.scatter(stars_g_i[::x0],stars_i_K[::x0],alpha=a, color='blue')
ax4.scatter(qsr_g_i[::x1],qsr_i_K[::x1], alpha=a, color='green')
ax4.scatter(gal_g_i[::x2],gal_i_K[::x2], alpha=a, color='red')
ax4.set_xlabel('$g - i$',fontsize=18)
ax4.set_ylabel('$i - K_{s}$',fontsize=18)
ax4.tick_params(labelsize=16)
ax4.axis([-1.0,4.0,0.0,5.0])

# Draw lines on plot 2
xx=np.array([[0,3],[.65,1.4]])
ax2.plot(xx[0,:],xx[1,:],color='black')
ax2.vlines(x=0.6, ymin=.8, ymax=5, color='black')
ax2.vlines(x=0.0, ymin=0, ymax=0.65, color='black')

# Draw lines on plot 4
yy=np.array([[0,3],[1.5,4.3]])
ax4.plot(yy[0,:],yy[1,:],color='black')
ax4.vlines(x=0.6, ymin=2.1, ymax=5, color='black')
ax4.vlines(x=0.0, ymin=0, ymax=1.5, color='black')

# Annotate plot 2
ax2.annotate('QSOs',(-0.8,2.25), fontsize=16, color='g')
ax2.annotate('Galaxies',(1.0,2.0), fontsize=16, color='r')
ax2.annotate('Stars',(2.0,0.5), fontsize=16, color='b')

# Annotate plot 4
ax4.annotate('QSOs',(-0.8,2.25), fontsize=16, color='g')
ax4.annotate('Galaxies',(2.2,4.4), fontsize=16, color='r')
ax4.annotate('Stars',(1.5,1.4), fontsize=16, color='b')
```
